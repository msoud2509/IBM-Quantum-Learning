# Unitary Operations

```python
from qiskit.quantum_info import Statevector, Operator
from qiskit.circuit.library import PhaseGate
import numpy as np
```

Like stochastic matrices are for classical representations, unitary matrices are the mathematical objects that represent quantum operations. 

A unitary matrix is a complex square matrix $U$ that satisfies the condition:
$$
U^\dagger U = U U^\dagger = I
$$

Where $U^\dagger$ is the conjugate transpose of $U$, and $I$ is the identity matrix.

Additionally, unitary matrices have the property that their inverse is equal to their conjugate transpose:
$$
U^{-1} = U^\dagger
$$

Unitary matrices have the powerful property that, when multiplied with a vector, they preserve the euclidean norm of the resultant vector:
$$\| U \ket{\psi} \| = \| \ket{\psi} \|$$

In other words, a quantum state vector multiplied by a unitary matrix results in an output state vector that is also a valid quantum state vector.

## Types of Unitary Operations
This notebook covers the following unitary matrices:
- Phase Operations ($P_\theta$)
    - Note that the common S & T gates are special cases of the phase operation, where $S = P_{\pi/2}$ and $T = P_{\pi/4}$.
- Pauli Matrices ($I, X, Y, Z$)
- Hadamard Matrix ($H$)

## Phase Operations ($P_\theta$)
A phase operation is a unitary operation that applies a relative phase shift to the quantum state (for an analysis of global vs relative phases and why global phases are irrelevant, visit this [notebook???](???)). It is represented by the matrix:
$$P_\theta = \begin{bmatrix}
1 & 0 \\
0 & e^{i\theta}
\end{bmatrix}$$
For any real number $\theta$.

The relative shift does a shift in the complex plane, so after applying a phase shift, the probabilities of measurement remain the same, but the relation between the amplitudes of the state can change, which can affect interference patterns in quantum algorithms (more information [here???](???)).

Note that the phase matrix is always unitary regardless of the value of $\theta$.


```python
phi = Statevector([(1 + 2j) / 3, (-2)/3])
print("Original State:")
display(phi.draw('latex', prefix="|\\phi\\rangle = "))

# notice that theta of 0 is just the identity operation (described below in paul matrices section), 
# so it should not change the state at all
theta = 0
phase_op = Operator(PhaseGate(theta))
phi_0 = phi.evolve(phase_op)

theta = np.pi / 4
phase_op = Operator(PhaseGate(theta))
phi_1 = phi.evolve(phase_op)

theta = np.pi / 2
phase_op = Operator(PhaseGate(theta))
phi_2 = phi.evolve(phase_op)

# notice that theta of pi is just the Z gate (described below in paul matrices section)
theta = np.pi
phase_op = Operator(PhaseGate(theta))
phi_3 = phi.evolve(phase_op)

print("State after applying phase shift of 0:")
display(phi_0.draw('latex', prefix="|\\phi_0\\rangle = "))
print("State after applying phase shift of π/4:")
display(phi_1.draw('latex', prefix="|\\phi_1\\rangle = "))
print("State after applying phase shift of π/2 (S-gate):")
display(phi_2.draw('latex', prefix="|\\phi_2\\rangle = "))
print("State after applying phase shift of π (Z-gate):")
display(phi_3.draw('latex', prefix="|\\phi_3\\rangle = "))
```

## Pauli Operations
The 4 pauli matrices are:
$$I = 
\begin{bmatrix}
1 & 0 \\
0 & 1
\end{bmatrix}, \quad
X = 
\begin{bmatrix}
0 & 1 \\
1 & 0
\end{bmatrix}, \quad
Y = 
\begin{bmatrix}0 & -i \\
i & 0
\end{bmatrix}, \quad
Z = 
\begin{bmatrix}1 & 0 \\
0 & -1
\end{bmatrix}
$$
*Note*: $X, Y, Z$ are often referred to as $\sigma_x, \sigma_y, \sigma_z$. Additionally, these 3 matrices correspond to the single-qubit, $X, Y, Z$ gates you will often see in quantum circuits.

In summary the Pauli matrices have the below behavior:
- $I (P_0)$ leaves the state unchanged (therefore, you will generally not see identity gates in quantum circuits involving single qubit systems).
- $X$ does a NOT operation, flipping the amplitudes of the $|0⟩$ and $|1⟩$ states.
- $Y$ also does a NOT operation, but with an additional phase flip (i.e., it multiplies the amplitude of the $|1⟩$ state by -1j).
    - For illustration, if $|\psi\rangle = \alpha|0\rangle + \beta|1\rangle$, then $Y|\psi\rangle = \beta(-i|0\rangle) + \alpha(i|1\rangle)$.
- $Z (P_\pi)$ applies a phase (sign) flip to the $|1⟩$ state, leaving the $|0⟩$ state unchanged.

The behaviors are further explored below:


```python
# Identity Matrix (I)
I = Operator([[1, 0], [0, 1]])

ket0 = Statevector.from_label('0')
ket1 = Statevector.from_label('1')
plus = Statevector.from_label('+')

I_ket0 = ket0.evolve(I)
I_ket1 = ket1.evolve(I)
I_plus = plus.evolve(I)

print("Identity Operation Results:")
print("\nI|0⟩ =")
display(I_ket0.draw('latex'))

print("\nI|1⟩ =")
display(I_ket1.draw('latex'))

print("\nI|+⟩ =")
display(I_plus.draw('latex'))

# notice the results are the same as the original states, as expected
```

```python
# Pauli-X Matrix (X-gate)
X = Operator([[0, 1], [1, 0]])

ket0 = Statevector.from_label('0')
ket1 = Statevector.from_label('1')
minus = Statevector.from_label('-')
phi = Statevector([(1 + 2j) / 3, (-2)/3])

print("Original States:")
display(ket0.draw('latex', prefix="|0⟩ = "), )
display(ket1.draw('latex', prefix="|1⟩ = "), )
display(minus.draw('latex', prefix="|-⟩ = "), )
display(phi.draw('latex', prefix="|\\phi⟩ = "), )

X_ket0 = ket0.evolve(X)
X_ket1 = ket1.evolve(X)
X_minus = minus.evolve(X)
X_phi = phi.evolve(X)

print("\nResults after applying Pauli-X (NOT) Operation:")
print("\nX|0⟩ =")
display(X_ket0.draw('latex'))
print("\nX|1⟩ =")
display(X_ket1.draw('latex'))
print("\nX|-⟩ =")
display(X_minus.draw('latex'))
print("\nX|φ⟩ =")
display(X_phi.draw('latex'))

# notice the X-gate flips the amplitudes of the |0⟩ and |1⟩ states, 
# for example the minus state now has the negative on ket0 instead of ket1
```

```python
# Pauli-Y Matrix (Y-gate)
Y = Operator([[0, -1j], [1j, 0]])

ket0 = Statevector.from_label('0')
ket1 = Statevector.from_label('1')
plus = Statevector.from_label('-')
minus = Statevector.from_label('+')

Y_ket0 = ket0.evolve(Y)
Y_ket1 = ket1.evolve(Y)
Y_plus = plus.evolve(Y)
Y_minus = minus.evolve(Y)

print("Results after applying Pauli-Y Operation:")
print("|0⟩ * Y =")
display(Y_ket0.draw('latex'))
print("|1⟩ * Y =")
display(Y_ket1.draw('latex'))
print("|+⟩ * Y =")
display(Y_plus.draw('latex'))
print("|-⟩ * Y =")
display(Y_minus.draw('latex'))
```

```python
# Pauli-Z Matrix (Z-gate)
Z = Operator([[1, 0], [0, -1]])

plus = Statevector.from_label('+')

Z_plus = plus.evolve(Z)

print("Results after applying Pauli-Z Operation:")
print("Z|+⟩ =")
display(Z_plus.draw('latex'))

# notice the only difference is that the sign of the ket1 amplitude is flipped
# since Z is a phase operation, the actual probabilities of measurement remain the same
```

## Hadamard Operation (H-gate)
$$
H = \frac{1}{\sqrt{2}} \begin{bmatrix}
1 & 1 \\
1 & -1
\end{bmatrix}
$$

The Hadamard operation is another unitary matrix that has the special properties of converting the standard basis states $|0⟩$ and $|1⟩$ into equal superpositions of each other (the $|+⟩$ and $|-⟩$ states).

```python
# Hadamard Matrix (H-gate)
H = Operator([[1/np.sqrt(2), 1/np.sqrt(2)], [1/np.sqrt(2), -1/np.sqrt(2)]])

ket0 = Statevector.from_label('0')
ket1 = Statevector.from_label('1')

H_ket0 = ket0.evolve(H)
H_ket1 = ket1.evolve(H)

print("Results after applying Hadamard Operation to |0⟩ and |1⟩ (results should be plus and minus states, respectively):")
display(H_ket0.draw('latex', prefix="H|0⟩ = "))
display(H_ket1.draw('latex', prefix="H|1⟩ = "))

print("Using the Hadamard operation again on the results should give us back the original states:")
H_H_ket0 = H_ket0.evolve(H)
H_H_ket1 = H_ket1.evolve(H)
display(H_H_ket0.draw('latex', prefix="H(H|0⟩) = "))
display(H_H_ket1.draw('latex', prefix="H(H|1⟩) = "))
```

Notice that the Hadamard matrix allows us to discrimate between the $|+⟩$ and $|-⟩$ states, or more generally, allows us to discrimiate phase changes, as we can apply the Hadamard operation to a state, and then measure in the standard basis to determine if the state was $|+⟩$ or $|-⟩$.

## Compositions of Unitary Operations
Unitary operations can be composed together by multiplying their corresponding matrices. The resulting matrix will also be unitary, and the composition of the operations will have the combined effect of the individual operations.

```python
H = Operator([[1/np.sqrt(2), 1/np.sqrt(2)], [1/np.sqrt(2), -1/np.sqrt(2)]])
S = Operator(PhaseGate(np.pi/2))

# R = H * S * H
R = H.compose(S).compose(H)
print("R=HSH:")
display(R.draw('latex'))



```

Notice that $R$ from above is unitary, and interestingly $(HSH)^2$ is equivalent to the $X-gate$, or the NOT operation, meaning that $HSH$ is a square root of the NOT operation, so applying $HSH$ twice is equivalent to applying the NOT operation once. This feature is not possible in classical computing.
