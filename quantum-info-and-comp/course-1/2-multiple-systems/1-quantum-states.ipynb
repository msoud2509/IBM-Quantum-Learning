# Quantum States

```python
from qiskit.quantum_info import Statevector
import numpy as np
```

The Cartesian product is the set of all possible ordered pairs of elements from two sets. For example, if we have two sets $A = {1, 2}$ and $B = {x, y}$, the Cartesian product $A \times B$ would be $\{(1, x), (1, y), (2, x), (2, y)\}$.

We can apply this concept to quantum states. If we have two quantum systems, we can use the tensor product to describe the combined state of the two systems. For example, if we have Ket-0 ($|0\rangle$) and Ket-1 ($|1\rangle$) as two quantum states, we can represent a combined state by using the tensor product:
$$|0\rangle \otimes |1\rangle$$

Which is equivalent in notation to the below:
$$|0\rangle \otimes |1\rangle = |0\rangle|1\rangle = |01\rangle$$

We call the two individual quantum states (in this case, $|0\rangle$ and $|1\rangle$) to be in a product state.

Note that just like in singular systems, a combined quantum state also needs to have a euclidean norm of 1 (therefore any tensor product between two valid quantum states will also be a valid quantum state):


```python
c00 = 1 / np.sqrt(2)
c01 = -1 / np.sqrt(6)
c10 = 1j / np.sqrt(6)
c11 = 1 / np.sqrt(6)

phi = Statevector([c00, c01, c10, c11]) # valid

display(phi.draw('latex'))
print("Is \\phi a valid quantum state?", phi.is_valid())

c00 = 1
c01 = 1
c10 = 0
c11 = 0
psi = Statevector([c00, c01, c10, c11]) # not valid

display(psi.draw('latex'))
print("Is \\psi a valid quantum state?", psi.is_valid())
```

```python
# tensor product demo
plus = Statevector([1 / np.sqrt(2), 1 / np.sqrt(2)])
minus = Statevector([1 / np.sqrt(2), -1 / np.sqrt(2)])

# both of the below syntax is valid for computing the tensor product of two quantum states
phi = plus.tensor(minus)
phi = plus ^ minus

display(phi.draw('latex', prefix="|+\\rangle \\otimes |-\\rangle = "))
```
