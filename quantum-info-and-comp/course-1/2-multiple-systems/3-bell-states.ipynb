# Bell States

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from numpy import sqrt
```

The 4 bell states are as follows:
$$|\phi^+\rangle = \frac{|00\rangle + |11\rangle}{\sqrt{2}}$$
$$|\phi^-\rangle = \frac{|00\rangle - |11\rangle}{\sqrt{2}}$$
$$|\psi^+\rangle = \frac{|01\rangle + |10\rangle}{\sqrt{2}}$$
$$|\psi^-\rangle = \frac{|01\rangle - |10\rangle}{\sqrt{2}}$$

These 4 states form a basis for the 2-qubit state space, and are all special cases of entangled states.

```python
# generate |00> and |11> states from bell basis
phi_plus = Statevector([1 / sqrt(2), 0, 0, 1 / sqrt(2)])
phi_minus = Statevector([1 / sqrt(2), 0, 0, -1 / sqrt(2)])

output = (1 / sqrt(2)) * (phi_plus + phi_minus)
display(output.draw('latex', prefix="\\frac{1}{\\sqrt{2}}(|\\phi^+\\rangle + |\\phi^-\\rangle) = "))

output = (1 / sqrt(2)) * (phi_plus - phi_minus)
display(output.draw('latex', prefix="\\frac{1}{\\sqrt{2}}(|\\phi^+\\rangle - |\\phi^-\\rangle) = "))
```

## Creating Bell States via Quantum Circuits
Quantum circuits are touched on [later in this course???](???) (and it's probably worth checking out first to understand the behavior of the C-NOT gates), but it's cool to see how we can achieve the bell basis using unitary operations on pairs of qubits.

Note that qubits in `qiskit.QuantumCircuit` are initialized to the $|0\rangle$ state by default.

```python
# Reaching the Phi+ Bell State
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
display(qc.draw("mpl"))

state = Statevector.from_instruction(qc)
display(state.draw('latex', prefix="|\\Phi^+\\rangle = "))
```

```python
# Reaching the Phi- Bell State
qc = QuantumCircuit(2)
qc.x(0)
qc.h(0)
qc.cx(0, 1)
display(qc.draw("mpl"))

state = Statevector.from_instruction(qc)
display(state.draw('latex', prefix="|\\Phi^-\\rangle = "))
```

```python
# Reaching the Psi+ Bell State
qc = QuantumCircuit(2)
qc.x(1)
qc.h(0)
qc.cx(0, 1)
display(qc.draw("mpl"))

state = Statevector.from_instruction(qc)
display(state.draw('latex', prefix="|\\Psi^+\\rangle = "))
```

```python
# Reaching the Psi- Bell State
qc = QuantumCircuit(2)
qc.x(1)
qc.h(0)
qc.cx(0, 1)
qc.z(1)
display(qc.draw("mpl"))

state = Statevector.from_instruction(qc)
display(state.draw('latex', prefix="|\\Psi^-\\rangle = "))
```
