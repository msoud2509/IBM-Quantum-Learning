# Entangled States

```python
from qiskit.quantum_info import Statevector
from numpy import sqrt
```

Not all quantum state vectors are product states, meaning that they cannot be written as a tensor product of two individual quantum states, and hence, the individual quantum states cannot be described independently of each other. Such states are called entangled states. For example, the state vector below is an entangled state:
$$|\phi\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle)$$

Below is a proof that the above state is not a product state (based off of [IBM's proof](https://quantum.cloud.ibm.com/learning/en/courses/basics-of-quantum-information/multiple-systems/quantum-information#entangled-states) with a little more explanation):

We will do a proof by contradiction, so we will assume that $|\phi\rangle$ is a product state, meaning that it can be written as a tensor product of two individual quantum states:
$$|\phi\rangle = |\phi_1\rangle \otimes |\phi_2\rangle$$

In order to get a coefficient of a component of a quantum state, we can take the inner product of the quantum state with the basis vector corresponding to that component. In this case, to get the coefficient of $|01\rangle$ in $|\phi\rangle$, we can take the inner product of $|\phi\rangle$ with $|01\rangle$, which should equal 0 based on the definition of $|\phi\rangle$ (verification provided in python cell below):
$$\langle 01|\phi\rangle = \langle 0 \otimes 1|\phi_1 \otimes \phi_2\rangle = 0$$

Since tensor products and inner products are distributive, we can rewrite the above as:
$$\langle 0|\phi_1\rangle\langle 1|\phi_2\rangle = 0$$

This means that either $\langle 0|\phi_1\rangle = 0$ or $\langle 1|\phi_2\rangle = 0$ (or both). 


However, based on the same process we just did, we get the following equations for the coefficients of $|00\rangle$ and $|11\rangle$ in $|\phi\rangle$:
$$\langle 00|\phi\rangle = \langle 0|\phi_1\rangle\langle 0|\phi_2\rangle = \frac{1}{\sqrt{2}}$$
$$\langle 11|\phi\rangle = \langle 1|\phi_1\rangle\langle 1|\phi_2\rangle = \frac{1}{\sqrt{2}}$$


Meaning that neither $\langle 0|\phi_1\rangle$ nor $\langle 1|\phi_2\rangle$ can be 0, which is a contradiction. Therefore, our original assumption that $|\phi\rangle$ is a product state must be false, and hence, $|\phi\rangle$ is an entangled state.

```python
phi = Statevector([1 / sqrt(2), 0, 0, 1 / sqrt(2)])

basis_01 = Statevector.from_label('01')

inner_product = basis_01.inner(phi)
print(f"The inner product <01|phi> is: {inner_product}")
```
