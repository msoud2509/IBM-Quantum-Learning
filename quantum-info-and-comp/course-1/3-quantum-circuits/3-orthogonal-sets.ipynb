# Orthogonal & Orthonormal Sets

```python
import numpy as np
from qiskit.quantum_info import Statevector
```

2 vectors $\ket{\psi}$ and $\ket{\phi}$ are said to be *orthogonal* if their inner product is zero:
$$\langle\psi|\phi\rangle = 0$$

A set of vectors $\{\ket{\psi_1}, \ket{\psi_2}, \cdots, \ket{\psi_n}\}$ is said to be an *orthogonal set* if every pair of distinct vectors in the set is orthogonal:
$$\langle\psi_i|\psi_j\rangle = 0 \quad \text{ for all } i \neq j$$

A set of vectors $\{\ket{\psi_1}, \ket{\psi_2}, \cdots, \ket{\psi_n}\}$ is said to be an *orthonormal set* if it is an orthogonal set and every vector in the set has unit euclidean norm:
$$\langle\psi_i|\psi_j\rangle = \begin{cases}
1 & \text{ if } i = j \\
0 & \text{ if } i \neq j
\end{cases}$$

```python
def is_orthonormal_basis(vectors: list[Statevector]) -> bool:
    # Check if the vectors are orthogonal
    for i in range(len(vectors)):
        for j in range(i + 1, len(vectors)):
            if vectors[i].inner(vectors[j]) != 0:
                return False
    
    for vector in vectors:
        if not vector.is_valid(): # checks if the vector is normalized
            return False
            
    return True

# Should output true
phi = Statevector([1/np.sqrt(2), 1/np.sqrt(2)])
psi = Statevector([1/np.sqrt(2), -1/np.sqrt(2)])
vectors = [phi, psi]
print("Vector set:")
for v in vectors:
    display(v.draw("latex"))
print("Is the set orthonormal?", is_orthonormal_basis(vectors))

# Should output false
phi = Statevector([1/np.sqrt(2), 1/np.sqrt(2)])
psi = Statevector([1/2, 1/2])
vectors = [phi, psi]
print("Vector set:")
for v in vectors:    
    display(v.draw("latex"))
print("Is the set orthonormal?", is_orthonormal_basis(vectors))
```

Additionally, a set of vectors $\{\ket{\psi_1}, \ket{\psi_2}, \cdots, \ket{\psi_n}\}$ is said to be an *orthonormal basis* if it is an orthonormal set and $n$ equals the dimension of the vector space.

## Orthonormal Sets & Unitary Matrices
The rows of any unitary matrix form an orthonormal set, and the columns of any unitary matrix also form an orthonormal set.
