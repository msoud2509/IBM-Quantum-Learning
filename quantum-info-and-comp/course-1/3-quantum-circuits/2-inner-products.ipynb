# Inner Products

Given a two vectors $\ket{\psi}$ and $\ket{\phi}$, the inner product is represented as $\langle\psi|\phi\rangle$ and is calculated as the matrix product of the bra vector $\langle\psi|$ and the ket vector $\ket{\phi}$ (where the bra vector $\langle\psi|$ is the conjugate transpose of the ket vector $\ket{\psi}$). For more information about the definition of the inner product, see the corresponding [IBM section](https://quantum.cloud.ibm.com/learning/en/courses/basics-of-quantum-information/quantum-circuits/inner-products-and-projections#inner-products).

## Some properties of inner products
1. **Euclidean norm & Positive definiteness**:
    Given a ket vector $\ket{\psi}$, the inner product $\langle\psi|\psi\rangle$ equals the euclidean norm of the vector $\ket{\psi}$ squared:
    $$\langle\psi|\psi\rangle = \|\ket{\psi}\|^2$$
    $$\|\ket{\psi}\| = \sqrt{\langle\psi|\psi\rangle}$$

    Because of this property, $\langle\psi|\psi\rangle > 0$ for all non-zero vectors $\ket{\psi}$, this is often called the *positive definiteness* property of inner products.

2. **Conjugate symmetry**: Given 2 ket vectors $\ket{\psi}$ and $\ket{\phi}$, the inner product $\langle\psi|\phi\rangle$ is the complex conjugate of the inner product $\langle\phi|\psi\rangle$:
$$\langle\psi|\phi\rangle = \overline{\langle\phi|\psi\rangle}$$

3. **Linearity in the 2nd argument & conjugate linearity in the 1st argument**:
    Given a ket vector $\ket{\psi}$ and 2 complex scalars $\alpha$ and $\beta$, where $\ket{\psi} = \alpha\ket{\psi_1} + \beta\ket{\psi_2}$, the inner product $\langle\phi|\psi\rangle$ can be expanded as:

    $$\langle\phi|\psi\rangle = \alpha\langle\phi|\psi_1\rangle + \beta\langle\phi|\psi_2\rangle$$

    Hence, the inner product is linear in the 2nd argument. On the other hand, the inner product is conjugate linear in the 1st argument, which means that:

    $$\langle\psi|\phi\rangle = \overline{\alpha}\langle\psi_1|\phi\rangle + \overline{\beta}\langle\psi_2|\phi\rangle$$

4. **Cauchy-Schwarz inequality**: Given 2 ket vectors $\ket{\psi}$ and $\ket{\phi}$, the inner product $\langle\psi|\phi\rangle$ is always less than or equal to the product of the euclidean norms of the two vectors:
$$|\langle\psi|\phi\rangle| \leq \|\ket{\psi}\| \cdot \|\ket{\phi}\|$$
