# Projections & Projective Measurements

```python
from qiskit.quantum_info import Operator, Statevector
import numpy as np
```

A square matrix $P$ is called a *projection* if it satisfies the following two properties:
1. **Idempotence**: $P^2 = P$
2. **Hermitian**: $P^\dagger = P$

```python
def conjugate_transpose(matrix: Operator) -> Operator:
    return matrix.conjugate().transpose()

def is_idempotent(matrix: Operator) -> bool:
    return matrix.compose(matrix) == matrix

def is_projection(matrix: Operator) -> bool:
    return is_idempotent(matrix) and matrix == conjugate_transpose(matrix)
```

```python
S = Operator.from_label('S')
display(S.draw('latex', prefix='S = '))
display(conjugate_transpose(S).draw('latex', prefix='S^\\dagger = '))

display(S.compose(S).draw('latex', prefix='S^2 = '))
print(f"Is S idempotent? {is_idempotent(S)}")

print(f"Is S a projection? {is_projection(S)}")

I = Operator.from_label('I')
display(I.draw('latex', prefix='I = '))
display(conjugate_transpose(I).draw('latex', prefix='I^\\dagger = '))

display(I.compose(I).draw('latex', prefix='I^2 = '))
print(f"Is I idempotent? {is_idempotent(I)}")
print(f"Is I a projection? {is_projection(I)}")
```

A square matrix $P$ is a projection if it can be expressed as $P = \ket{\psi}\bra{\psi}$ for some unit vector $\ket{\psi}$. So this means that any quantum state vector $\ket{\psi}$ can be associated with a projection matrix $P = \ket{\psi}\bra{\psi}$.

```python
plus = Statevector.from_label('+')
display(plus.draw('latex', prefix='+ = '))
square_plus = plus.to_operator() # to_operator performs the outer product |+><+|
display(square_plus.draw('latex', prefix='|+\\rangle\\langle+| = '))
print(f"Is |+><+| a projection? {is_projection(square_plus)}")

psi = Statevector([(1 + 2j) / 3, (-2) / 3])
display(psi.draw('latex', prefix='\\psi = '))
square_psi = psi.to_operator()
display(square_psi.draw('latex', prefix='|\\psi\\rangle\\langle\\psi| = '))
print(f"Is |psi><psi| a projection? {is_projection(square_psi)}")
```

## Projective Measurements
A set of projection matrices $\{P_0...P_n\}$ is a *projective measurement* if the sum of the matrices in the set equal the identity matrix:
$$ \sum_{i=0}^n P_i = I $$

Or more generally, a set of projective matrices can be defined as any $a$ within a set $\Sigma$:
$$ \sum_{a \in \Sigma} P_a = I $$

And for each $a \in \Sigma$, we can represent the probability of obtaining outcome $a$ when measuring a quantum state $\ket{\psi}$ as:
$$ Pr(a) = ||P_a \ket{\psi}||^2 $$

And the resultant state if you obtain outcome $a$ can be represented as:
$$ \ket{\psi_a} = \frac{P_a \ket{\psi}}{\sqrt{Pr(a)}} = \frac{P_a \ket{\psi}}{||P_a \ket{\psi}||} $$

As an example, in the 2-qubit system, the set of projective matrices $\{P_1, P_0\}$ can be constructed by the Bell states:
$$ P_0 = |\phi^+\rangle\langle\phi^+| + |\phi^-\rangle\langle\phi^-| + |\psi^+\rangle\langle\psi^+| \quad \text{and} \quad P_1 = |\psi^-\rangle\langle\psi^-| $$

```python
phi_plus = Statevector([1 / np.sqrt(2), 0, 0, 1 / np.sqrt(2)])
phi_minus = Statevector([1 / np.sqrt(2), 0, 0, -1 / np.sqrt(2)])
psi_plus = Statevector([0, 1 / np.sqrt(2), 1 / np.sqrt(2), 0])
psi_minus = Statevector([0, 1 / np.sqrt(2), -1 / np.sqrt(2), 0])

proj0 = phi_plus.to_operator() + phi_minus.to_operator() + psi_plus.to_operator()
proj1 = psi_minus.to_operator()
display(proj0.draw('latex', prefix='P_0 = '))
display(proj1.draw('latex', prefix='P_1 = '))

qubit0 = Statevector([(1 + 2j) / 3, (-2) / 3])
qubit1 = Statevector.from_label('1')
phi = qubit0.tensor(qubit1)
display(phi.draw('latex', prefix='|\\phi\\rangle = '))
pr_1 = (phi.expectation_value(proj1)).real
print(f"Probability of outcome 1: {pr_1}")
state_after_measurement = phi.evolve(proj1) / np.linalg.norm(phi.evolve(proj1).data)
display(state_after_measurement.draw('latex', prefix='State after measurement: '))
```
